```python
# https://fr.mathworks.com/help/wavelet/ref/scal2frq.html
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
%matplotlib inline
```

```python
def get_dilated_morlet(M, w=5.0, s=1., dt=.00625):
    x       = np.linspace(-M/2,M/2,M+1) * dt #np.linspace(-2*np.pi, 2*np.pi, M) * s
    output  = np.exp(1j * w * x * s)
    output -= np.exp(-0.5 * (w**2))
    output *= np.exp(-0.5 * ((x*s)**2)) * np.pi**(-0.25)
    output *= np.power(1 + np.exp(-w**2) - 2 * np.exp(-3./4 * w**2) , -1./2)
    output *= s
    return [x, output]

def get_dilated_morlet_fourier(M, w=5.0, s=1., unif=False):
    #w0      = w*s
    if unif==True:
        x       = np.linspace(0, .5 * np.pi * 2, M)/s
    else:
        x       = np.linspace(0, 50 * np.pi * 2, M)/s
    output  = np.exp(-1./2*(w - x)**2)
    output -= np.exp(-0.5 * (w**2)) * np.exp(-0.5 * (x**2))
    output *= np.power(1 + np.exp(-w**2) - 2 * np.exp(-3./4 * w**2) , -1./2) * np.pi**(-0.25)
    if unif==True:
        return [np.linspace(0, .5 * np.pi * 2, M), output]
    else:
        return [np.linspace(0, 50 * np.pi * 2, M), output]
```

```python
# fourier of wavelets
nb_filter   = 44
s_list_f    = 2**(np.arange(nb_filter)/15.)
w_list_f    = np.ones(nb_filter)*2*np.pi*5
for i in range(len(w_list_f)):
    x, morlet_d = get_dilated_morlet_fourier(M=1000, w=w_list_f[i], s=s_list_f[i], unif=False)
    plt.plot(x/(np.pi*2), morlet_d/np.max(morlet_d)); plt.hold(True)
plt.xlabel('frequencies')
plt.ylabel('normalized wavelet')
plt.title('wavelet in fourier')
frequencies = w_list_f*s_list_f/(2*np.pi)
print(frequencies)
```

```python
# filter bank
nb_filter   = 44
M           = 50
dt          = 0.00625
s_list_f    = 2**(np.arange(nb_filter)/15.)
w_list_f    = np.ones(nb_filter)*2*np.pi*5
wavelets    = np.zeros([nb_filter, 2, M])
for i in range(len(w_list_f)):
    x, morlet_d = get_dilated_morlet(M=M-1, w=w_list_f[i],s=s_list_f[i], dt=dt)
    wavelets[i,0] = np.real(morlet_d)
    wavelets[i,1] = np.imag(morlet_d)
print('wavelet bank shape is : {0}'.format(wavelets.shape))
```
